```python
import requests
import json

f = open('agoda_query.json')
data = json.load(f)

response = requests.post(
    'https://www.agoda.com/graphql/search',
    headers={
        'Content-Type': 'application/json',
        'Ag-Language-Locale': 'zh-tw'
    },
    json=data
)

print(response.json())
```
