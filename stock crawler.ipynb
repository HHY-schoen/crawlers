- yfinance 的資料來源是 Yahoo Finance API，可以將網路上的資料給下載下來給程式使用

```python
# pip install yfinance
import yfinance as yf

df = yf.Ticker("2330.TW").history(period="max")
df = df.reset_index()      # 重設索引為0
df
```

```python
import pandas as pd

df = df.round({'Close':2})
df['Date'] = df['Date'].dt.date      # .dt.date 只留下日期
df = df.loc[:,['Date','Close']].set_index('Date')
df
```

- 將 dataframe 存成 csv 檔

```python
df.to_csv(f"stock.csv")
```
