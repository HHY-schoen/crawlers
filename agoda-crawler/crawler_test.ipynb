- 到 Headers 查看請求的方式與參數
> 網址是 https://www.agoda.com/graphql/search
，使用 POST 方法、需要帶 payload、採用 GraphQL 查詢語言。

因為請求需要夾帶的 JSON 太長，另外存在一個檔案(agoda_query.json)內再讀取進來。

```python
import requests
import json

f = open('agoda_query.json')
data = json.load(f)

response = requests.post(
    'https://www.agoda.com/graphql/search',
    headers={
        'Content-Type': 'application/json',
        'Ag-Language-Locale': 'zh-tw'
    },
    json=data
)

print(response.json())
```

```python
# 將回傳結果儲存在 JSON 檔案
with open('agoda_result.json', 'w') as fp:
    json.dump(response.json(), fp)
```
